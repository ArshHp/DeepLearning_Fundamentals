Text Summarization 

Use Case: Summarized the customer feedback or the CPE support work into a summarized document for training corpus. 

Usage: 
  
  Internal to R&D to build the summarized KB documents. 

  As a service to the SAAS user for consumption

```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization")
```

```python
!pip install gradio
```

```python
import gradio as gr
import pandas as pd
```

```python
def Summarize(text):
  ret_text = summarizer(text)
  return ret_text[0]['summary_text']
  
  
```

```python
My_Text = """
    The document needs to be updated with a command (given below)  which checks whether the existing postgres server certificate contains subjectAlternativeName(SAN) or not before instructing the users to Update PostgreSQL database server certificate after OMT upgrade.

openssl x509 -text -in <Postgres server.crt> -noout | grep "Subject Alternative Name" 

https://docs.microfocus.com/doc/Containerized_Operations_Bridge/2022.05/Preupgradereqs

https://docs.microfocus.com/doc/Containerized_Operations_Bridge/2022.05/EnableTLSInPostgrSQL

Update PostgreSQL database server certificate after OMT upgrade

The usage of server certificates without subjectAlternativeName(SAN) extension is not supported in 2022.05. Matching the hostname by relying on checking the “Common Name” field is not supported. Any certificates not having a SAN need to be updated or recreated before upgrading to 2022.05 and the same must be used during the upgrade. For more details on the certificate requirement details, See Enable TLS in PostgreSQL. 
"""
Summarize(My_Text)

  
```

```python
# creating the interface
app = gr.Interface(fn=Summarize, inputs = ['text'], 
                   outputs = 'textbox', 
                   title = 'Document Summarizer', 
                   theme= "darkpeach", 
                   description = 'Please paste your text content for the summarization',
                   allow_flagging = "auto")
```

```python
# launching the app
app.launch(inline=False, debug=True, share=True)
```
